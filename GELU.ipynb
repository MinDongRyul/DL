```python
import numpy as np
```

```python
np.random.randn(1, 1000)
```

```python
value = np.linspace(-4, 3, 1000)
```

```python
def gelu(x):
    return x / (1 + np.exp(-1.702 * x))
```

```python
def gelu2(x):
    return 0.5 * x * (1 + np.tanh(np.sqrt(2 / np.pi) * (x + 0.44715 * x ** 3)))
```

```python
def elu(x):
    alpha = 1
    return np.where(x > 0, x, alpha * (np.exp(x) - 1))
```

```python
def relu(x):
    return np.where(x > 0, x, 0)
```

```python
value2 = gelu(value)
value3 = gelu2(value)
value4 = relu(value)
value5 = elu(value)
```

```python
import matplotlib.pyplot as plt
```

```python
# plt.plot(value, label = '-4.0 ~ 3.0')
plt.plot(value4, label = 'relu', color = 'b')
plt.plot(value2, label = 'gelu(silu)', color = 'green')
plt.plot(value3, label = 'gelu(tanh)', color = 'black')
plt.plot(value5, label = 'elu', color = 'r')
plt.legend()
plt.title('GELU activation')
plt.show();
```
