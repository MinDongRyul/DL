```python
import numpy as np
import math
from math import factorial as fact

import matplotlib.pyplot as plt
```

```python
value = np.arange(-4, 4, 0.05,  dtype=np.float32)
value.shape
```

```python
def maclaurin(x, type_):
    if type_ == 2:
        return 1 - 1/fact(2) * (x ** 2)
    elif type_ == 4:
        return 1 - 1/fact(2) * (x ** 2) + 1/fact(4) * (x ** 4)
    elif type_ == 6:
        return 1 - 1/fact(2) * (x ** 2) + 1/fact(4) * (x ** 4) - 1/fact(6) * (x ** 6)
    elif type_ == 8:
        return 1 - 1/fact(2) * (x ** 2) + 1/fact(4) * (x ** 4) - 1/fact(6) * (x ** 6) + 1/fact(8) * (x ** 8)
```

```python
value2 = maclaurin(value, 2)
value3 = maclaurin(value, 4)
value4 = maclaurin(value, 6)
value5 = maclaurin(value, 8)
```

```python
plt.figure(figsize=(8, 6))

plt.plot(value, np.cos(value), label='cosx')
plt.plot(value, value2, label='1 - 1/2! * (x^2)')
plt.plot(value, value3, label='1 - 1/2! * (x^2) + 1/4! * (x^4)')
plt.plot(value, value4, label='1 - 1/2! * (x^2) + 1/4! * (x^4) - 1/6! * (x^6)')
plt.plot(value, value5, label='1 - 1/2! * (x^2) + 1/4! * (x^4) - 1/6! * (x^6) + 1/8! * (x^8)')

plt.legend()

# 레이아웃 조정
plt.tight_layout()
plt.title('Maclaurin Series')
plt.show()
```
