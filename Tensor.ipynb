```python
# 딥러닝으로 간단한 수학문제 풀어보기
# 키 = [170, 180, 175, 160]
# 신발 = [260, 270, 265, 255]

# ??? = a 170 + b (신발 = a * 키 + b)
키 = 170
신발 = 260

# 신발 = 키 * a + b

a = tf.Variable(0.1) # 변수 만드는 문법
b = tf.Variable(0.2)

def 손실함수(): # loss fuction 손실함수 만들기(실제값과 예측값의 차이를 구하기)
    예측값 = 키 * a + b
    return tf.square(260 - 예측값) # 음수방지를 위해 제곱해서 리턴
                    #실제값 - 예측값
opt = tf.keras.optimizers.Adam(learning_rate=0.1) # a,b 경사하강법으로 구하기

for _ in range(300): # 원하는 만큼 경사하강 반복
    opt.minimize(손실함수, var_list=[a, b]) 
    print(a.numpy(), b.numpy())
# var_list=[a, b] : 경사하강법으로 업데이트할 weight Variable 목록 삽입
```

```python
import tensorflow as tf
import pandas as pd
import numpy as np

data = pd.read_csv('/Users/민동률/DL/data/gpascore.csv')
# data.isnull().sum() 빈칸 확인
data = data.dropna()
# data = data.fillna()

y데이터 = data['admit'].values
x데이터 = []

for i, rows in data.iterrows():
    x데이터.append([rows['gre'], rows['gpa'], rows['rank']])

# 1. 딥러닝 model 디자인하기
# Sequential쓰면 신경망 레이어들 쉽게 만들어줌
model = tf.keras.models.Sequential([
    tf.keras.layers.Dense(64, activation='tanh'),  
    # 64 : hidden layer에서 node의 갯수
    # 레이어 갯수/ 노드 갯수는 결과가 잘 나올 때까지 실험으로 파악
    # 레이어에 activation function(활성화함수) 넣기
    # ex : sigmoid, tanh, relu, softmax, leakyRelu
    tf.keras.layers.Dense(128, activation='tanh'),
    tf.keras.layers.Dense(1, activation='sigmoid'), 
    # 마지막 출력레이어(확률 한개를 예측할 때도 1개)
    # 예측결과(여기서는 0~1사이의 확률/ 이유는 sigmoid가 0~1사이로 바꿔주기 때문)를 
    # 뱉어야함
]) # <= 딥러닝 모델

# 2. model compile
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
# optimizer : 가중치 수정(learning late 같은 걸 조절?)
# ex : adam, adagrad, adadelta, rmsprop, sgd
# loss : 손실 함수
# binary_crossentropy : 결과가 0과 1사이의 분류/확률문제에서 씀
# metrics : 성능 측정 지표 정하기

# 3. model 학습(fit) 시키기
history = model.fit(np.array(x데이터), np.array(y데이터), epochs=1000) # numpy array 또는 tf tensor로 넣어줘야함
# x데이터 : 학습 데이터
# y데이터 : 실제 정답
# epochs : 몇번 학습 시킬지

# 예측
예측값 = model.predict([[750, 3.7, 3], [400, 2.2, 1]])
print(예측값)
```

```python
# 훈련의 정확도와 손실 그래프 그리기
import matplotlib.pyplot as plt
import numpy as np

def history_plot(history_dict):
    acc = history_dict['accuracy']
#     val_acc = history_dict['val_accuracy']
    loss = history_dict['loss']
#     val_loss = history_dict['val_loss']

    epochs = np.arange(1, len(acc)+1)
    plt.plot(epochs, acc,'bo', label='train')
#     plt.plot(epochs, val_acc, label='val')
    plt.title('Accuracy')
    plt.legend()
    plt.show()

    plt.plot(epochs, loss, 'bo', label='train')
#     plt.plot(epochs, val_loss, label='val')
    plt.title('Loss')
    plt.legend()
    plt.show()

history_plot(history.history)
```
